$x_{ac} \geq y_{bc}d_{ab} \quad \forall a\in A, b\in B, c\in C$

```python
#Función que calcula el costo de realizar tomas ese día
def costo_tomas(matriz_datos, tomas, c=1):
    if not tomas:
        return 0
    # Selecciona solo las columnas de las tomas
    submatriz = np.array(matriz_datos)[:, tomas]
    # Para cada actor (fila), verifica si participa en alguna toma seleccionada
    actores_que_asisten = np.any(submatriz, axis=1)
    # Suma el costo para cada actor que asiste
    return np.sum(actores_que_asisten) * c
```

```python
def optimizar_tomas(tomas_restantes, particion_actual, costo_actual, matriz_datos, combinaciones_tomas,
                    c, mejor_costo, mejor_particion,id = str(0) ):
    #print(mejor_costo)
    if not tomas_restantes:
        if costo_actual < mejor_costo:
            return costo_actual, [list(grupo) for grupo in particion_actual]
        else:
            return mejor_costo, mejor_particion
    #Tomamos un grupo de tomas de la lista en cada iteración
    combinaciones_validas = [g for g in combinaciones_tomas if all(t in tomas_restantes for t in g)]
    contador = 0
    now = time.time()
    last_print_time = [now, now]
    for grupo in combinaciones_tomas:
        contador +=1
        #print(id,contador)
        #Imprimo el número de iteraciones cada minuto
        if now - last_print_time[0] >= 120:
            print(f"Iteraciones: {contador}, tiempo transcurrido: {int((now - last_print_time[1]) // 60)} minutos")
            last_print_time[0] = now
        if all(t in tomas_restantes for t in grupo):
            #Calculamos el costo por grabar ese día
            nuevo_costo = costo_actual + costo_tomas(matriz_datos, grupo, c)
            if nuevo_costo >= mejor_costo:
                continue
            nuevas_tomas_restantes = [t for t in tomas_restantes if t not in grupo]
            #print("nuevas_tomas_restantes",nuevas_tomas_restantes)
            particion_actual.append(grupo)
            costo_resultado, particion_resultado = optimizar_tomas(nuevas_tomas_restantes, particion_actual,
                                                          nuevo_costo, matriz_datos,
                                                        combinaciones_tomas,
                                                          c, mejor_costo, mejor_particion, id = id+" "+str(contador)
                                                                  )
            if costo_resultado < mejor_costo:
                mejor_costo, mejor_particion = costo_resultado, particion_resultado
                print(f"Nuevo mejor costo: {mejor_costo}, partición: {mejor_particion}")
            particion_actual.pop()


    return mejor_costo, mejor_particion
```

```python
 """
    Deja solo combinaciones donde:
      - No aparece ninguna toma_junta,
      - O aparecen todas las tomas_juntas juntas.
    Elimina combinaciones con solo una o dos de las tomas_juntas.
    """
def filtrar_combinaciones_por_juntas(combinaciones_tomas, tomas_juntas):

    tomas_juntas_set = set(tomas_juntas)
    resultado = []
    for grupo in combinaciones_tomas:
        grupo_set = set(grupo)
        inter = grupo_set.intersection(tomas_juntas_set)
        if len(inter) == 0 or inter == tomas_juntas_set:
            resultado.append(grupo)
    return resultado
```

$O(n) = O(\sum_{k=1}^m\binom{n}{k}\cdot \sum_{k=1}^m\binom{n-m}{k}\sum_{k=1}^m\binom{n-2m}{k}\cdot...)=O(m*n!)$

** (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta**

Respuesta

Generamos el siguiente algoritmo:

1.- Consideramos las tomas en que coinciden los mismos actores. Se obtiene una lista

2.- Ordenamos por la lista más larga probamos las otras listas para obtener el grupo de m tomas con el menor costo.

3.- En el paso 2 podemos obtener multiples soluciones con el mismo costo para cada una calculamos la nueva lista de tomas restantes y repetimos el algoritmo


Ejemplo consideramos 3 actores y 4 tomas, con maximo 3 tomas por día:

El costo por actor es c

Actor 1, 2 Tomas: 1, 2

Actor 1 Toma 3

Actor 3 Toma 4

Obtenemos:

Actor 1,2  Toma: 1, 2, 3 Costo 2*c

Actor 3 Toma: 4  Costo 1*c




```python
from itertools import combinations
import copy

def tomas_con_mismos_actores(matriz):
    grupos = {}
    T = len(matriz[0])
    for t in range(T):
        actors = frozenset(i for i,row in enumerate(matriz) if row[t])
        grupos.setdefault(actors, []).append(t)
    return list(grupos.values())

def costo_tomas(matriz, tomas, c=1):
    actors = set(a for a in range(len(matriz)) if any(matriz[a][t] for t in tomas))
    return len(actors) * c

def nuevo_optimizar_tomas(matriz, max_tomas_por_dia, c=1):
    clusters = tomas_con_mismos_actores(matriz)
    size = [len(cl) for cl in clusters]
    best_global = (float('inf'), None)

    def construir_horario(rem_clusters):
        """Construye un voraz puro sobre el conjunto rem_clusters."""
        rem = set(rem_clusters)
        sched = []
        total_cost = 0
        while rem:
            # 1) cluster(s) de mayor tamaño
            max_sz = max(size[i] for i in rem)
            cand_max = [i for i in rem if size[i]==max_sz]
            # 2) entre ellos, escoge los de menor cost/size
            min_ratio = float('inf')
            for i in cand_max:
                cost_i = costo_tomas(matriz, clusters[i], c)
                ratio = cost_i / size[i]
                min_ratio = min(min_ratio, ratio)
            cand_ties = [i for i in cand_max
                         if abs(costo_tomas(matriz, clusters[i], c)/size[i] - min_ratio)<1e-9]

            # si sólo hay uno, seguimos voraz normal
            best_i = cand_ties[0]
            day_tomas = clusters[best_i].copy()
            day_cost  = costo_tomas(matriz, day_tomas, c)
            rem.remove(best_i)

            # rellenar día
            while True:
                candidata, best_r = None, float('inf')
                for j in rem:
                    if len(day_tomas)+size[j] > max_tomas_por_dia:
                        continue
                    new_tomas = day_tomas + clusters[j]
                    new_cost  = costo_tomas(matriz, new_tomas, c)
                    delta = new_cost - day_cost
                    ratio = delta / size[j]
                    if ratio < best_r:
                        best_r, candidata = ratio, j
                if candidata is None:
                    break
                rem.remove(candidata)
                day_tomas += clusters[candidata]
                day_cost  = costo_tomas(matriz, day_tomas, c)

            sched.append(day_tomas)
            total_cost += day_cost

        return total_cost, sched

    # Explora viejos empatados  ---
    all_clusters = set(range(len(clusters)))
    # calcula empates del día 1
    max_sz = max(size)
    cand_max = [i for i in all_clusters if size[i]==max_sz]
    best_ratio = float('inf')
    # filtro por ratio dentro de los de max_sz
    for i in cand_max:
        r = costo_tomas(matriz, clusters[i], c)/size[i]
        best_ratio = min(best_ratio, r)
    ties_day1 = [i for i in cand_max
                 if abs(costo_tomas(matriz, clusters[i], c)/size[i] - best_ratio)<1e-9]

    # prueba cada empate como inicio
    for first in ties_day1:
        rem = all_clusters.copy()
        # fuerza elegir `first` en el día 1
        rem.remove(first)
        day1 = clusters[first].copy()
        cost1 = costo_tomas(matriz, day1, c)
        # rellena día1 igual que voraz
        while True:
            cand2, best_r2 = None, float('inf')
            for j in rem:
                if len(day1)+size[j] > max_tomas_por_dia: continue
                new_t = day1 + clusters[j]
                new_c = costo_tomas(matriz, new_t, c)
                dr = (new_c - cost1)/size[j]
                if dr < best_r2:
                    best_r2, cand2 = dr, j
            if cand2 is None: break
            rem.remove(cand2)
            day1 += clusters[cand2]
            cost1 = costo_tomas(matriz, day1, c)

        # construye el resto con voraz puro
        cost_rest, sched_rest = construir_horario(rem)
        total = cost1 + cost_rest
        sched = [day1] + sched_rest

        if total < best_global[0]:
            best_global = (total, sched)

    return best_global

```
