## Torres de Hanoi con Divide y vencerás

Reglas del juego

1- Tienes tres varillas (o postes): A, B y C.

2- En una de las varillas (por ejemplo, A) hay n discos, ordenados de mayor a menor (el más grande abajo).

3- El objetivo es mover todos los discos desde la varilla A hasta la varilla C.

4- Solo puedes mover un disco a la vez.

5- Nunca puedes colocar un disco grande encima de uno más pequeño.

```python
#Considerando que el algoritmo resulve el problema clásico, 
#añado condiciones para que el input (desde, hasta) considere valores de 1 a 3
def Torres_Hanoi(N, desde = 1, hasta = 3):
    if N ==1 :
        print("Lleva la ficha " ,desde , " hasta ", hasta )
    else:
        Torres_Hanoi(N-1, desde, 6-desde-hasta )
        print("Lleva la ficha " ,desde , " hasta ", hasta )
        Torres_Hanoi(N-1, 6-desde-hasta  , hasta )
        
def Input_Hanoi(N, desde = 1, hasta = 3):
    if not desde in [1,2,3]:
        print("La variable 'desde' esta mal definida")
    elif not hasta in [1,2,3]:
        print("La variable 'hasta' esta mal definida")
    else:
        return(Torres_Hanoi(N, desde = 1, hasta = 3))
    

Input_Hanoi(3, 1 , 3)
```

Alternativa de código para las torres de hanoi

```python
def hanoi(n, origen, auxiliar, destino):
    if n == 1:
        print(f"Mover disco 1 de {origen} a {destino}")
    else:
        hanoi(n-1, origen, destino, auxiliar)
        print(f"Mover disco {n} de {origen} a {destino}")
        hanoi(n-1, auxiliar, origen, destino)
```

# Sucesión_de_Fibonacci

https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci

Calculo del termino n-simo de la suscesión de Fibonacci

```python
#en esta lista guardamos los valores de fibonacci
def Fibonacci(N:int):
    if N < 2:
        return 1
    else:
        x =Fibonacci(N-1)+Fibonacci(N-2)
        print(x)
        return x
```

```python
Fibonacci(5)
```

Se puede notar como el algoritmo recalcula números ya procesados como el 2 y 3 

Alternativa para verificar los números calculados

```python
def Fibonacci(n, memo=None):
    if memo is None:
        memo = {}
    if n < 2:
        return 1
    if n not in memo:
        memo[n] = Fibonacci(n-1, memo) + Fibonacci(n-2, memo)
        print(memo)
    return memo[n]
```

```python
Fibonacci(5)
```

## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
    SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
    ValorAcumulado = 0
    for i,valor in enumerate(SM):
        monedas =  (N-ValorAcumulado)//valor
        SOLUCION[i] = monedas
        ValorAcumulado = ValorAcumulado + monedas*valor

        if ValorAcumulado == N:
              return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
    n = len(S)
    for x in range(n):
        print("")
        for i in range(n):
            if S[i] == x+1:
                print(" X " , end="")
            else:
                print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
    for i in range(etapa+1):
        #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
        if SOLUCION.count(SOLUCION[i]) > 1:
            return False

    #Verifica las diagonales
        for j in range(i+1, etapa +1 ):
          #print("Comprobando diagonal de " + str(i) + " y " + str(j))
          if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
    return True



def reinas(N, solucion=[], etapa=0):
    if len(solucion) == 0:
        solucion=[0 for i in range(N)]

    for i in range(1, N+1):
        solucion[etapa] = i

        if es_prometedora(solucion, etapa):
            if etapa == N-1:
                print(solucion)
                escribe(solucion)
                print()
            else:
                reinas(N, solucion, etapa+1)
        else:
            None
        solucion[etapa] = 0

reinas(8)
```

## Viaje por el rio. Programación dinámica

```python
infinito = float('inf')


TARIFAS = [
[0,5,4,3,infinito,infinito,infinito],
[infinito,0,infinito,2,3,infinito,11],
[infinito,infinito, 0,1,infinito,4,10],
[infinito,infinito, infinito, 0,5,6,9],
[infinito,infinito, infinito,infinito,0,infinito,4],
[infinito,infinito, infinito,infinito,infinito,0,3],
[infinito,infinito,infinito,infinito,infinito,infinito,0]
]



################################################################
def Precios(TARIFAS):
################################################################
    #Total de Nodos
    N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
    PRECIOS = [ [infinito]*N for i in [infinito]*N]
    RUTA = [ [""]*N for i in [""]*N]

    for i in range(0,N-1):
        RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
        PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
        for j in range(i+1, N):
            MIN = TARIFAS[i][j]
            RUTA[i][j] = i

            for k in range(i, j):
                if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
                    MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
                    RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
                PRECIOS[i][j] = MIN
                
    return PRECIOS,RUTA
    ################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
    print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
    print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
    if desde == hasta:
        #print("Ir a :" + str(desde))
        return ""
    else:
        return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
