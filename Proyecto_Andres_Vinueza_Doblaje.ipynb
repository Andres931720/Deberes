<u>Conjuntos:</u>

$A$: Conjunto de actores

$T$: Conjunto de tomas

$C$: Conjunto de días tomados



<u>Variables:</u>

$x_{ac}\in\{0,1\}$: El actor $a$ grabada el día $c$

$y_{tc}\in\{0,1\}$: La toma $t$ graba el día $c$

<u>Parámetros:</u>

$c$: Costo por venir a grabar

$d_{at}$: Matriz con la información si participa un actor $a$ en la toma $t$

$n$: Número de tomas

Sabemos que $|C|<= |T|/n$

<u>*Función Objetivo:*</u>

Es un problema de minimización porque queremos reducir el costo total del doblaje

$$\min c\sum_{a \in A} \sum_{c\in C} x_{ac}$$

<u>Restricciones:</u>

Solo se tiene una toma al día

$\sum_{c\in C}y_{tc}=1\quad \forall t\in T$

Se puede aplicar máximo n tomas

$\sum_{c\in C}y_{tc}\leq n \quad \quad \forall c\in C$

Un actor $a$ acude el día $c$ si tiene que grabar ese día

$x_{ac} \geq y_{tc}d_{at} \quad \forall a\in A, t\in T, c\in C$

## **Diseña un algoritmo para resolver el problema por fuerza bruta**

Respuesta

Pasos (resumen)

1. $tomas = {0..T-1}$, $horario = []$, $costo=0$, $mejor\_costo = ∞$.

2. Si no hay tomas disponibles entonces tomamos la solución actual y costo actual.
    
    Entonces:
    
    $mejor\_costo = costo$
    
    return $horario$


3. Toma pivote $t_0 = min(tomas)$.

4. Genera todas las combinaciones de tomas factibles que no incluyen $t_0$:<br>
    Los candidatos son conjuntos de tomas con cardinalidad menor o igual a $n-1$.
    
    $$Combinaciones = \{X: t_0\notin X \land |X|\leq n-1 \}$$
    
5. Para cada conjunto de combinaciones lo unimos con $t_0$ y obtenemos el conjunto de tomas para el día correspondiente. Luego con las tomas del día calculamos el costo de las tomas de ese día. Finalmente actualizamos el conjunto de tomas y costo.

    Esto lo podemos representar así:

    $dia = \{t_0\}\cup X$, donde $X$ reprenta una combinación de $n-1$ tomas

    $c = calcular\_costo(dia)$

    $tomas = tomas - dia$
  
     $costo = costo + c$
     
     $horario = horario\cup \{dia\}$

6. repetir el algoritmo para las tomas restantes

Comentario:

- El algoritmo requiere calcular todas las combinaciones de tomas que pueden asignarse en un día. En la instancia actual, esto corresponde a los distintos conjuntos de 6 tomas. Sin embargo, si aumenta el número de tomas o de actores, este paso podría volverse computacionalmente inviable.

- Existe la posibilidad de visitar soluciones factibles más de una vez. Por ejemplo, al evaluar la solución en la que el conjunto de tomas {1,2,3,4,5,6} se asigna al día 1, también se podría evaluar otra solución en la que ese mismo conjunto se asigna al día 2. Dado que en el problema un mismo conjunto de tomas puede aplicarse en cualquier día sin afectar el costo, estas evaluaciones representan ramificaciones redundantes.

- El algoritmo admite instancias en las que el número total de tomas no es múltiplo del número de tomas permitidas por día. Por ejemplo, en una instancia con 30 tomas y un límite de 7 tomas por día, se podrían programar 7 tomas durante 4 días y luego completar el proceso con 2 tomas adicionales en un quinto día.

## **Calcula la complejidad del algoritmo por fuerza bruta**

Respuesta

Sea, $T = |T|$ , $A = |A|$ y $C = |C|$

Sea $L$ la función que cuenta las soluciones exploradas:
- Para el caso que la cantidad de tomas es múltiplo de la cantidad de tomas

\begin{align} L(T,n) &= \frac{T!}{(n!)^CC!} \\
                      \end{align}

- Para el caso general consideramos variables  $x_1,...,x_n$ que representan tal que $\sum i x_i=T$

$$L(T,n) = \sum_{k=1}^{\min{n,T}}\binom{T-1}{k-1}L(T-k,n) $$

Para calcular el tiempo multiplicamos por el número de actores y tomas:

Tiempo total ≈ (#hojas) × (trabajo por hoja) = $$O(L(T,A)\cdot A \cdot T)  $$

La siguiente función convierte los datos a una lista de conjuntos, cada conjunto son los actores participantes por toma

```python
 """
    Convierte un DataFrame (filas=tomas, columnas=actores) a:
      - takes_actors: list[set[int]] (o set[str] si keep_names=True)
      - A: número total de actores
      - actor_to_idx: dict {col_actor -> índice} (útil para interpretar resultados)

    Reglas para interpretar presencia del actor en la toma:
      - Si la columna es booleana: True/False tal cual.
      - Si es numérica: > 0 se considera presente.
      - En otros casos: string no vacío y no en {'0','false','no','nan'}.
    """
def df_tomas_actores(df: pd.DataFrame, keep_names: bool = False):

    # 1) Normaliza a booleano por columna
    #    (soporta mezcla de tipos)
    bcols = {}
    for col in df.columns:
        s = df[col]
        if pd.api.types.is_bool_dtype(s):
            bcols[col] = s.fillna(False)
        elif pd.api.types.is_numeric_dtype(s):
            bcols[col] = s.fillna(0).gt(0)            # > 0 => presente
        else:
            # texto: presente si no es vacío/”falso”
            s2 = s.fillna("").astype(str).str.strip().str.lower()
            bcols[col] = ~s2.isin({"", "0", "false", "no", "nan"})
    bdf = pd.DataFrame(bcols, index=df.index)

    # 2) Mapeo actor -> índice (0..A-1)
    actor_cols = list(bdf.columns)
    actor_to_idx = {actor: i for i, actor in enumerate(actor_cols)}
    A = len(actor_cols)

    # 3) Construir takes_actors
    takes_actors = []
    if keep_names:
        # sets con los NOMBRES de actores
        for _, row in bdf.iterrows():
            present = set(actor for actor, flag in row.items() if bool(flag))
            takes_actors.append(present)
    else:
        # sets con los ÍNDICES de actores (0..A-1)
        for _, row in bdf.iterrows():
            present = {actor_to_idx[actor] for actor, flag in row.items() if bool(flag)}
            takes_actors.append(present)

    return takes_actors, A, actor_to_idx
```

El siguiente es el algoritmo de fuerza bruta, toma la lista de conjuntos y retorna el mejor costo y mejor horario


Parámetros:

      - tomas_actores: lista donde tomas_actores[i] es un set de actores en la toma i
      - C: capacidad máxima de tomas por día
      - forzar_dias_completos: si True, cada día debe tener exactamente C tomas
      - imprimir_mejores: si True, imprime cada vez que mejora el costo
      - tiempo_max_seg: (float|None) límite de tiempo en segundos para cortar la búsqueda
                        Si es None, no hay límite.

```python
from itertools import combinations

from time import perf_counter
```

```python
# --- FUERZA BRUTA SIMPLE (SIN PODAS)  ---


def fuerza_bruta_simple(tomas_actores, C=6, forzar_dias_completos=False,
                        imprimir_mejores=True, tiempo_max_seg=None):

    T = len(tomas_actores)
    print("T:", T)
    print("C:", C)
    print(T % C)
    print(forzar_dias_completos)

    if forzar_dias_completos is False:
        forzar_dias_completos = (T % C <= 0.1)

    mejor_costo = float("inf")
    mejor_horario = None
    agotado_tiempo = False

    inicio = perf_counter()

    def tiempo_excedido():
        return (tiempo_max_seg is not None) and ((perf_counter() - inicio) >= tiempo_max_seg)

    def costo_dia(dia):
        actores = set()
        for t in dia:
            actores |= tomas_actores[t]
        return len(actores)

    def rec(restantes, horario, costo_actual):
        nonlocal mejor_costo, mejor_horario, agotado_tiempo

        # Chequeo de tiempo al entrar
        if tiempo_excedido():
            agotado_tiempo = True
            return

        if not restantes:
            if costo_actual < mejor_costo:
                mejor_costo = costo_actual
                mejor_horario = [d[:] for d in horario]
                if imprimir_mejores:
                    print(f"[NUEVO MEJOR] costo={mejor_costo}, días={len(mejor_horario)}")
            return

        # Pivote para evitar permutaciones equivalentes
        t0 = min(restantes)
        otros = sorted(restantes - {t0})

        if forzar_dias_completos:
            necesidad = C - 1
            if len(otros) < necesidad:
                return
            for combo in combinations(otros, necesidad):
                # Chequeo rápido de tiempo dentro del bucle
                if tiempo_excedido():
                    agotado_tiempo = True
                    return
                dia = [t0, *combo]
                c = costo_dia(dia)
                rec(restantes - set(dia), horario + [dia], costo_actual + c)
                if agotado_tiempo:
                    return
        else:
            max_k = min(C - 1, len(otros))
            for k in range(0, max_k + 1):
                for combo in combinations(otros, k):
                    if tiempo_excedido():
                        agotado_tiempo = True
                        return
                    dia = [t0, *combo]
                    c = costo_dia(dia)
                    rec(restantes - set(dia), horario + [dia], costo_actual + c)
                    if agotado_tiempo:
                        return

    rec(set(range(T)), [], 0)
    return mejor_costo, mejor_horario, agotado_tiempo
```

El algoritmo podría tomar muchos años en resolver el problema por lo que se ejecutará para un tiempo y se mostrará la solución.

```python
best_cost, best_sched, _ = fuerza_bruta_simple(tomas_actores[0], C=6, forzar_dias_completos=True, tiempo_max_seg = 100)
print("\n== SOLUCIÓN ==")
print("Costo mínimo actor-día:", best_cost)
for d, tomas in enumerate(best_sched, 1):
    print(f"Día {d}: tomas {sorted(tomas)}  (#{len(tomas)})")
```

## **(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta**

Respuesta

Se aplica una alternativa con programación dinámica. La PD almacena las soluciones intermedias (en una tabla o caché) y las reutiliza, esto evita cálculos redundantes.

El algoritmo sugerido anteriormente recalcula las opciones de tomas por día alargando la busqueda.

### Cómo funciona paso a paso

1. Estado inicial: todas las tomas están pendientes.
Ejemplo: (0, 1, 2, 3, 4, 5) si hay 6 tomas.

2. Programación dinámica (DP):

    - El algoritmo mira un conjunto de tomas pendientes.

    - Selecciona un grupo de K tomas para formar un día.

    - Calcula el costo de ese día.

    - Llama recursivamente para resolver las tomas restantes.

    - Guarda el mejor resultado encontrado (mínimo costo).

3. Evitar repeticiones:

    - Se usa memoización (lru_cache) para no recalcular lo mismo dos veces.

    - Siempre se obliga a que el grupo elegido incluya la toma más pequeña que queda pendiente, así se evita contar varias veces combinaciones que son equivalentes pero en distinto orden.

4. Reconstrucción:

    Una vez calculado el costo mínimo, se van siguiendo las elecciones guardadas para armar la agenda completa:

    Día 1 → estas K tomas,

    Día 2 → otras K tomas,

    … y así hasta cubrir todo.

### Notación y entrada

Sea el conjunto de tomas $T=\{0,1,\dots,n-1\}$.  
Tamaño fijo del día: $K$ (p.ej. $K=6$).  
Cada día es un subconjunto $D\subseteq T$ con $|D|=K$ y sin solaparse con otros días.

- Actores por toma: $\mathrm{act}(t)\subseteq \mathcal{A}$.  
- Costo de un día $D$:

$$
c(D) \;=\; c_{\text{fijo}} \;+\;
\Bigl|\;\bigcup_{t\in D}\mathrm{act}(t)\;\Bigr|\cdot c_{\text{actor}}.
$$

Objetivo: partir $T$ en días $D_1,\dots,D_m$ (con $|D_i|=K$ y $\biguplus_i D_i=T$) minimizando

$$
C^\star \;=\; \sum_{i=1}^{m} c(D_i).
$$

---

### Algoritmo (DP + búsqueda, días de tamaño fijo $K$)

**Estado.**  
$\mathrm{pend}\subseteq T$: tomas pendientes.  
$\mathrm{horario}$: lista de días fijados.  
$C$: costo acumulado.  
$C_{\min}$: mejor costo global.

**Inicialización.**  
$\mathrm{pend}\leftarrow T,\;\mathrm{horario}\leftarrow[\,],\;C\leftarrow 0,\;C_{\min}\leftarrow+\infty.$

$
\boxed{
\begin{array}{l}
\textbf{Procedimiento } \mathrm{Resolver}(\mathrm{pend}, \mathrm{horario}, C):\\[2pt]
\quad \textbf{si } |\mathrm{pend}| = 0 \textbf{ entonces }\\
\qquad C_{\text{mejor}} \leftarrow \min(C_{\text{mejor}},\, C);\;\; \textbf{return}.\\[4pt]
\quad \text{(\emph{Propagación de obligados})}\\
\quad \text{mientras exista } t\in \mathrm{pend} \text{ que pertenezca a exactamente un candidato } D \text{ con } D\subseteq \mathrm{pend},\, |D|=K:\\
\qquad \mathrm{horario} \leftarrow \mathrm{horario} \mathbin{+\!\!=} [D];\;\; C \leftarrow C + c(D);\;\; \mathrm{pend} \leftarrow \mathrm{pend}\setminus D.\\
\qquad \textbf{si } |\mathrm{pend}| \bmod K \neq 0 \textbf{ entonces return } \;(\text{rama imposible}).\\[4pt]
\quad \text{(\emph{Cota inferior: branch\&bound})}\\
\quad r \leftarrow \left\lceil \dfrac{|\mathrm{pend}|}{K} \right\rceil;\;\;
\mathrm{LB} \leftarrow C + r \cdot \mathrm{lb\_dia}(\mathrm{pend}).\\
\quad \textbf{si } \mathrm{LB} \ge C_{\text{mejor}} \textbf{ entonces return } \;(\text{poda}).\\[6pt]
\quad \text{Elige un pivote } p = \min(\mathrm{pend}).\\
\quad \text{Para cada candidato } D \subseteq \mathrm{pend} \text{ con } |D|=K \text{ y } p\in D \text{ (ordenado por } c(D)\text{ asc.)}:\\
\qquad \mathrm{Resolver}(\mathrm{pend}\setminus D,\; \mathrm{horario}\mathbin{+\!\!=}[D],\; C+c(D)).\\
\end{array}
}
$

**Definición de la cota inferior.**  

$$
\operatorname{lb\_dia}(\mathrm{pend})
\;=\; c_{\text{fijo}}
+ \min_{\substack{D\subseteq\mathrm{pend}\\|D|=K}}
\Bigl|\;\bigcup_{t\in D}\mathrm{act}(t)\;\Bigr|\cdot c_{\text{actor}}.
$$



### Comentarios:

El algoritmo de fuerza bruta se podía mejorar aplicando algunas podas.

Se probaron las siguientes:

    - Anti-duplicados: evitar la visita de soluciones que son equivalentes. Por ejemplo, si tenemos un conjunto de tomas $T_1$ para el día 1 no tenemos un beneficio en colocarlas en el día 2 u otro.
    - Factibilidad:
    Si una toma tiene 1 único día posible ⇒ ese día es forzoso (se elige).
    Si alguna toma tiene 0 opciones ⇒ poda inmediata.


### Aplicación del algoritmo por programación dinámica

    Resuelve el problema asumiendo que TODOS los días tienen EXACTAMENTE k_tomas_por_dia tomas.
    Si 'tiempo_max_seg' se excede, devuelve el MEJOR resultado conocido hasta ese momento.

    Devuelve: (costo, agenda)
      * agenda: lista de días, cada día es una tupla de índices de tomas

```python
from functools import lru_cache
from itertools import combinations
import time

# ======================================================
# Función de costo por día
# ======================================================
def costo_dia_uniforme_actores(tomas_del_dia, tomas_actores, costo_actor_uniforme=1.0, costo_fijo_dia=0.0):
    actores = set()
    for t in tomas_del_dia:
        actores |= tomas_actores[t]
    return costo_fijo_dia + costo_actor_uniforme * len(actores)

# ======================================================
# Algoritmo principal
# ======================================================
def resolver_doblaje_pd(
    tomas_actores,
    k_tomas_por_dia=6,
    costo_actor_uniforme=1.0,
    costo_fijo_dia=0.0,
    verbose=True,
    tiempo_max_seg=None
):

    n = len(tomas_actores)
    if n == 0:
        return 0.0, []
    if n % k_tomas_por_dia != 0:
        raise ValueError(f"Con días llenos: n={n} debe ser múltiplo de K={k_tomas_por_dia}.")

    # ---------------------------
    # Reloj y excepción de corte
    # ---------------------------
    t0 = time.time()

    class _StopPorTiempo(Exception):
        pass

    def _tiempo_excedido():
        return (tiempo_max_seg is not None) and (time.time() - t0 >= tiempo_max_seg)

    # Conjunto inicial de tomas pendientes como tupla ordenada (0..n-1)
    full_state = tuple(range(n))

    # Para reconstrucción (guardamos la mejor elección por estado)
    mejor_eleccion = {}

    # Envoltorio de costo que usa los parámetros proporcionados
    def costo_de_dia(tomas_del_dia):
        return costo_dia_uniforme_actores(
            tomas_del_dia, tomas_actores,
            costo_actor_uniforme=costo_actor_uniforme,
            costo_fijo_dia=costo_fijo_dia
        )

    # -----------------------------------------
    # Incumbente inicial (siempre factible)
    # -----------------------------------------
    # Agrupa en bloques consecutivos de tamaño K
    agenda_incumbente = []
    for i in range(0, n, k_tomas_por_dia):
        agenda_incumbente.append(tuple(range(i, i + k_tomas_por_dia)))

    costo_incumbente = sum(costo_de_dia(d) for d in agenda_incumbente)

    mejor_global_costo = costo_incumbente
    mejor_global_agenda = [tuple(d) for d in agenda_incumbente]  # copia

    # -----------------------------------------
    # Para reconstruir agenda óptima desde 'mejor_eleccion'
    # -----------------------------------------
    def reconstruir_agenda(desde_estado):
        agenda = []
        estado = desde_estado
        while len(estado) > 0:
            dia = mejor_eleccion.get(estado, None)
            if dia is None:
                # Si falta alguna elección (p.ej. por corte de tiempo), deja de reconstruir
                break
            agenda.append(tuple(dia))
            estado = tuple(x for x in estado if x not in dia)
        return agenda

    # DP con memoización (estado: tupla ORDENADA de tomas pendientes)
    @lru_cache(maxsize=None)
    def dp(estado):
        nonlocal mejor_global_costo, mejor_global_agenda

        if _tiempo_excedido():
            raise _StopPorTiempo()

        m = len(estado)
        if m == 0:
            return 0.0
        if m == k_tomas_por_dia:
            # un solo día: costo directo
            costo = costo_de_dia(estado)
            mejor_eleccion[estado] = estado  # por claridad
            return costo

        # Heurística para romper simetrías: siempre incluir el menor índice pendiente
        pivote = estado[0]

        mejor = float("inf")
        mejor_dia = None

        # Generar solo días de tamaño K que incluyan el pivote
        for dia in combinations(estado, k_tomas_por_dia):
            if pivote not in dia:
                continue

            if _tiempo_excedido():
                raise _StopPorTiempo()

            # Estado restante = estado sin los elementos de 'dia'
            resto = tuple(x for x in estado if x not in dia)

            costo_total = costo_de_dia(dia) + dp(resto)
            if costo_total < mejor:
                mejor = costo_total
                mejor_dia = dia

                # Guardamos la elección provisional para poder reconstruir
                mejor_eleccion[estado] = mejor_dia

                # Si la mejora ocurre en el estado completo, actualizamos incumbente global
                if estado == full_state and mejor < mejor_global_costo:
                    agenda_actual = reconstruir_agenda(full_state)
                    # Solo actualiza si la reconstrucción está completa
                    if len(agenda_actual) * k_tomas_por_dia == n:
                        mejor_global_costo = mejor
                        mejor_global_agenda = [tuple(d) for d in agenda_actual]
                        if verbose:
                            print(f"Nuevo mejor costo TOTAL: {mejor_global_costo:.2f} (día elegido {list(mejor_dia)})")

        # Guardamos para reconstrucción final
        mejor_eleccion[estado] = mejor_dia
        return mejor

    # Ejecutar DP en el estado completo, con manejo de tiempo
    try:
        costo_minimo = dp(full_state)
        # Reconstrucción de la agenda óptima
        agenda = reconstruir_agenda(full_state)
        if len(agenda) * k_tomas_por_dia != n:
            return mejor_global_costo, mejor_global_agenda
        return costo_minimo, agenda

    except _StopPorTiempo:
        # Tiempo agotado: devolvemos la mejor solución conocida hasta ahora
        if verbose:
            print("[PARADA POR TIEMPO] Devolviendo mejor solución conocida.")
        return mejor_global_costo, mejor_global_agenda

```

```python
K = 6

print("Resolviendo problema con", len(tomas_actores[0]), "tomas y días de", K, "tomas...\n")
costo_min, agenda= resolver_doblaje_pd(
    tomas_actores[0],
    k_tomas_por_dia=K,
    costo_actor_uniforme=1.0,   # valor a pagar al actor por día
    costo_fijo_dia=0.0,         #  costo fijo por día
    verbose=True  ,              # imprime solo mejoras del costo global
    tiempo_max_seg =100          # tiempo maximo de solucion
)

print("\n=== Resultado final ===")
print("Costo mínimo:", round(costo_min, 2))
print("Agenda (cada tupla es un día con K tomas):", agenda)
```

### Complejidad del algoritmo

Este algoritmo trabaja con programación dinámica y memoización.  
El estado son las tomas pendientes y en cada paso se forman días de tamaño fijo $K$.  
Se obliga a incluir el pivote (la toma de menor índice) para no repetir combinaciones.

---

**Número de estados.**  
Solo aparecen estados cuyo tamaño es múltiplo de $K$:

$$
S(n,K)=\sum_{j=0}^{\lfloor n/K \rfloor} \binom{n}{jK}.
$$

Esto sigue siendo exponencial en $n$ cuando $K$ es fijo.

---

**Trabajo por estado.**  
En un estado con $s$ tomas pendientes, el algoritmo genera candidatos de tamaño $K$ que incluyen al pivote:

$$
\#\text{candidatos}(s,K)=\binom{s-1}{K-1}.
$$

---

**Tiempo total.**  
Cada estado hace ese número de intentos, por lo tanto:

$$
T(n,K)=
\sum_{s\in\{n,n-K,\dots,K\}}
\binom{n}{s}\cdot \binom{s-1}{K-1}.
$$

Esto es el peor caso, y es exponencial.

---

**Memoria.**  
Una entrada en la tabla por cada estado:

$$
\text{Espacio}=\Theta\!\left(\sum_{j}\binom{n}{jK}\right).
$$

---

### Caso aplicado: $n=30,\;K=6$

- Estados posibles:

$$
\binom{30}{0}=1,\;
\binom{30}{6}=593{,}775,\;
\binom{30}{12}=86{,}493{,}225,\;
\binom{30}{18}=86{,}493{,}225,\;
\binom{30}{24}=593{,}775,\;
\binom{30}{30}=1.
$$

- Total:

$$
S(30,6)=174{,}174{,}002 \;\; \text{estados}.
$$

- Transiciones (aprox):

$$
T(30,6)\;\approx\;5.95\times 10^{11}.
$$

---

### Resumen con mi forma de decirlo

El algoritmo al final es **exponencial**.  
Para $n=30$ y $K=6$ hay como $1.74\times 10^8$ estados y en la cota superior unas $5.95\times 10^{11}$ transiciones.  
En la práctica puede ser mucho menor si se mete poda con cota inferior y si se ordenan bien los candidatos, porque se corta el árbol antes.

 Genera un conjunto aleatorio de tomas y actores.
    Devuelve:
      - tomas_actores: lista donde tomas_actores[t] es un set de actores (0..num_actores-1)
      - dias_ref: partición en días consecutivos (tuplas de índices) de tamaño k_tomas_por_dia

    Parámetros clave:
      - num_actores: cantidad de actores disponibles (A)
      - k_tomas_por_dia: K tomas por día (tamaño fijo de cada día)
      - num_tomas: cantidad total de tomas (T), se exige T % K == 0
      - min/max_actores_por_toma: rango de actores por toma (incluyente)
      - semilla: fija el RNG
      - sesgo: define cómo se generan los pesos por actor
          * "uniforme": todos con la misma probabilidad
          * "mixto": pesos Beta (varios poco frecuentes, pocos frecuentes)
          * "estrella": pocos actores muy frecuentes y resto poco

```python
import random
from typing import List, Set, Tuple

def generar_tomas_actores_aleatorio(
    num_actores: int,
    k_tomas_por_dia: int,
    num_tomas: int,
    min_actores_por_toma: int ,
    max_actores_por_toma:  None,
    semilla:  None,
    sesgo: str = "mixto",  # "uniforme" | "mixto" | "estrella"
) -> Tuple[List[Set[int]], List[Tuple[int, ...]]]:
    if semilla is not None:
        random.seed(semilla)

    if num_actores <= 0 or k_tomas_por_dia <= 0 or num_tomas <= 0:
        raise ValueError("num_actores, k_tomas_por_dia y num_tomas deben ser > 0.")
    if num_tomas % k_tomas_por_dia != 0:
        raise ValueError("num_tomas debe ser múltiplo de k_tomas_por_dia.")
    if max_actores_por_toma is None:
        # Por defecto, ~10–25% de los actores por toma, acotado entre 1 y A
        max_actores_por_toma = max(1, min(num_actores, max(2, num_actores // 5)))
    if not (1 <= min_actores_por_toma <= max_actores_por_toma):
        raise ValueError("Se requiere 1 <= min_actores_por_toma <= max_actores_por_toma.")
    if max_actores_por_toma > num_actores:
        max_actores_por_toma = num_actores

    A = num_actores
    T = num_tomas
    K = k_tomas_por_dia
    actores = list(range(A))

    # --------------------------
    # Pesos por actor (sesgo)
    # --------------------------
    if sesgo == "uniforme":
        pesos = [1.0 for _ in actores]
    elif sesgo == "estrella":
        # Unos pocos muy frecuentes, resto poco frecuentes
        # p.ej., 2–4 estrellas
        n_star = max(2, min(4, A // 8 if A >= 8 else 2))
        pesos = [0.5 for _ in actores]
        estrellas = random.sample(actores, k=n_star)
        for i in estrellas:
            pesos[i] = 4.0
    else:  # "mixto" (por defecto): variedad suave
        # Beta(1.2, 3.0): mayoría baja frecuencia, algunos más altos
        pesos = [random.betavariate(1.2, 3.0) + 0.05 for _ in actores]

    # Asegurar pesos positivos
    eps = 1e-9
    pesos = [max(p, eps) for p in pesos]

    # --------------------------
    # Muestra ponderada SIN reemplazo (Efraimidis–Spirakis)
    # --------------------------
    def sample_sin_reemplazo_ponderado(k: int) -> Set[int]:
        """Selecciona k actores únicos ponderados por 'pesos'."""
        if k >= A:
            return set(actores)
        # Claves: r^(1/w) y elegimos las k mayores
        keys = [(random.random() ** (1.0 / w), i) for i, w in enumerate(pesos)]
        keys.sort(reverse=True)
        return {i for _, i in keys[:k]}

    # --------------------------
    # Construcción de tomas
    # --------------------------
    tomas_actores: List[Set[int]] = []
    for _ in range(T):
        s = random.randint(min_actores_por_toma, max_actores_por_toma)
        tomas_actores.append(sample_sin_reemplazo_ponderado(s))

    # Partición de referencia en días consecutivos (útil para pruebas)
    dias_ref = [tuple(range(i, i + K)) for i in range(0, T, K)]

    return tomas_actores, dias_ref
```

Adjunto como anexo una opción trabajada en el desarrollo donde se utilizan podas para resolver el problema. Aunque coloque "buenas" podas el algoritmo sigue calculando opciones de días ya visitadas.

```python
from itertools import combinations
from typing import List, Set, FrozenSet, Tuple, Optional, Dict
import math

# ============================================================
# Utilidades básicas
# ============================================================

def costo_dia(dia: Set[int], tomas_actores: List[Set[int]]) -> int:
    """Costo de un día = # de actores únicos en sus tomas."""
    actores = set()
    for t in dia:
        actores |= tomas_actores[t]
    return len(actores)


def generar_candidatos_completos(T: int, C: int) -> List[FrozenSet[int]]:
    """
    (Opcional) Genera TODOS los candidatos de día de tamaño C a partir de T tomas (0..T-1).
    Úsalo solo si no te pasan los candidatos ya construidos.
    """
    return [frozenset(d) for d in combinations(range(T), C)]


# ============================================================
# Backtracking con podas
# ============================================================

def planificar_desde_candidatos(
    tomas_actores: List[Set[int]],              # toma -> set de actores
    candidatos_dia: List[FrozenSet[int]],       # lista de candidatos (cada uno con C tomas)
    todas_las_tomas: Optional[Set[int]] = None, # universo a cubrir (por defecto, unión de candidatos)
    C: int = 6,
    permitir_empates_con_best: bool = True,     # si True, acepta costo == BEST
    imprimir_nuevos_mejores: bool = True,       # imprime solo cuando mejora la mejor solución
) -> Tuple[Optional[int], Optional[List[FrozenSet[int]]]]:
    """
    Elige un subconjunto disjunto de 'candidatos_dia' que cubra exactamente 'todas_las_tomas'
    y minimice el costo total (suma de #actores por día). Usa:
      - Regla del pivote (rompe-simetrías)
      - Viabilidad básica (toma con 0 opciones, unión insuficiente, conteo mínimo)
      - Propagación de obligados (unitarios)
      - Poda estricta por presupuesto con cotas inferiores (LB-A, LB-B, LB-C)
    """

    # --- Preparación de estructuras ---
    if todas_las_tomas is None:
        todas_las_tomas = set().union(*candidatos_dia) if candidatos_dia else set()

    # sets mutables para operaciones rápidas
    sets_cand: List[Set[int]] = [set(d) for d in candidatos_dia]
    costos: List[int] = [costo_dia(s, tomas_actores) for s in sets_cand]
    N = len(sets_cand)

    # índice inverso: toma -> {índices de candidatos que la contienen}
    indice_toma_a_candidatos: Dict[int, Set[int]] = {t: set() for t in todas_las_tomas}
    for i, S in enumerate(sets_cand):
        for t in (S & todas_las_tomas):
            indice_toma_a_candidatos[t].add(i)

    # --- Viabilidad global básica (rápida) ---
    # 1) Toda toma debe aparecer en al menos un candidato
    for t in todas_las_tomas:
        if not indice_toma_a_candidatos[t]:
            return None, None
    # 2) La unión de todos los candidatos debe cubrir el universo
    if not todas_las_tomas.issubset(set().union(*sets_cand) if sets_cand else set()):
        return None, None

    # --- Estado global del mejor ---
    mejor_costo = math.inf
    mejor_calendario: Optional[List[int]] = None

    # ============================================================
    # Funciones auxiliares para podas/cotas en el estado actual
    # ============================================================

    def viabilidad_basica(rem: Set[int], disponibles: Set[int]) -> bool:
        """Chequeos baratos de viabilidad en el estado (sin costo)."""
        if not rem:
            return True
        # conteo mínimo de candidatos (necesitas al menos r días)
        r = len(rem) // C
        if len(disponibles) < r:
            return False
        # cada toma debe tener al menos un candidato disponible que la contenga
        for t in rem:
            if not any((t in sets_cand[i]) for i in disponibles):
                return False
        # la unión de candidatos disponibles debe cubrir rem
        union_disp = set()
        for i in disponibles:
            union_disp |= sets_cand[i]
        if not rem.issubset(union_disp):
            return False
        return True

    def cota_inferior_estado(rem: Set[int], disponibles: Set[int]) -> int:
        """
        Cota inferior (admisible) del costo mínimo para completar el estado.
        Devuelve max(LB-A, LB-B, LB-C):
          - LB-A: suma de los r costos más bajos entre candidatos ⊆ rem.
          - LB-B: ceil( sum_t m(t) / C ), m(t)=min costo de cand (⊆ rem) que contiene t.
          - LB-C: sum_a ceil( k_a / C ), k_a=# tomas en rem donde aparece actor a.
        """
        if not rem:
            return 0
        r = len(rem) // C

        # A) costos más bajos (ignorando solapes)
        costos_validos = [costos[i] for i in disponibles if sets_cand[i].issubset(rem)]
        costos_validos.sort()
        if len(costos_validos) >= r:
            lb_a = sum(costos_validos[:r])
        else:
            # si faltan candidatos, LB-A subestima con lo que haya (viabilidad_basica ya debería cortar antes)
            lb_a = sum(costos_validos)

        # B) mínimo por toma
        suma_minimos = 0
        for t in rem:
            # menor costo entre disponibles que contengan t y estén dentro de rem
            mejor = math.inf
            for i in disponibles:
                if t in sets_cand[i] and sets_cand[i].issubset(rem):
                    if costos[i] < mejor:
                        mejor = costos[i]
            # si no hay ninguno, viabilidad_basica debería haber cortado; tratamos como inf
            if mejor is math.inf:
                return math.inf
            suma_minimos += mejor
        lb_b = math.ceil(suma_minimos / C)

        # C) por actor (mínimo de actor-día)
        # cuenta k_a = #tomas de rem donde aparece actor a
        conteo_por_actor: Dict[int, int] = {}
        for t in rem:
            for a in tomas_actores[t]:
                conteo_por_actor[a] = conteo_por_actor.get(a, 0) + 1
        lb_c = sum(math.ceil(k / C) for k in conteo_por_actor.values())

        return max(lb_a, lb_b, lb_c)

    # ============================================================
    # Propagación de obligados (unitarios)
    # ============================================================

    def propagar_obligados(rem: Set[int], disponibles: Set[int], elegidos: List[int]) -> Tuple[bool, Set[int], Set[int], List[int], int]:
        """
        Repite: si una toma aparece en EXACTAMENTE un candidato disponible, ese candidato es forzoso.
        Aplica todos los forzosos en cascada. Devuelve:
          (ok, rem', disponibles', elegidos_ext, costo_extra)
        """
        costo_extra = 0
        elegidos_ext: List[int] = []
        while True:
            # construye cobertura local t -> candidatos disponibles que contengan t
            cobertura: Dict[int, List[int]] = {}
            for t in rem:
                idxs = [i for i in disponibles if t in sets_cand[i]]
                if not idxs:
                    return False, rem, disponibles, elegidos_ext, costo_extra  # imposible
                cobertura[t] = idxs

            # candidatos forzosos (por tomas unitarias)
            forzosos = set()
            for t, idxs in cobertura.items():
                if len(idxs) == 1:
                    forzosos.add(idxs[0])

            if not forzosos:
                return True, rem, disponibles, elegidos_ext, costo_extra

            # aplica todos los forzosos (deduplicados)
            for ci in forzosos:
                if ci not in disponibles:
                    continue
                elegidos.append(ci)
                elegidos_ext.append(ci)
                costo_extra += costos[ci]
                cubiertas = sets_cand[ci]
                rem = rem - cubiertas
                # actualiza disponibles: solo los disjuntos sobreviven
                disponibles = {j for j in disponibles if sets_cand[j].isdisjoint(cubiertas)}

                if not rem:
                    return True, rem, disponibles, elegidos_ext, costo_extra

    # ============================================================
    # DFS principal
    # ============================================================

    def dfs(rem: Set[int], disponibles: Set[int], elegidos: List[int], costo_acum: int):
        nonlocal mejor_costo, mejor_calendario

        # poda por costo
        if costo_acum >= mejor_costo:
            return

        # viabilidad básica
        if not viabilidad_basica(rem, disponibles):
            return

        # propagación de obligados
        ok, rem, disponibles, extra_elegidos, costo_extra = propagar_obligados(rem, disponibles, elegidos)
        if not ok:
            # revertir efecto visual (no necesario en copia, pero por claridad)
            for _ in extra_elegidos:
                elegidos.pop()
            return

        costo_acum2 = costo_acum + costo_extra
        if not rem:
            # solución completa
            if costo_acum2 < mejor_costo or (permitir_empates_con_best and costo_acum2 == mejor_costo and mejor_calendario is None):
                mejor_costo = costo_acum2
                mejor_calendario = elegidos[:]
                if imprimir_nuevos_mejores:
                    #print(f"[NUEVO MEJOR] costo={mejor_costo} | días={(mejor_calendario)}")
                    print(f"[NUEVO MEJOR] costo={mejor_costo}")
            # deshacer elegidos extra antes de volver
            for _ in extra_elegidos:
                elegidos.pop()
            return

        # cota inferior del estado (estricta por presupuesto)
        lb_estado = cota_inferior_estado(rem, disponibles)
        if costo_acum2 + lb_estado > mejor_costo or (not permitir_empates_con_best and costo_acum2 + lb_estado >= mejor_costo):
            for _ in extra_elegidos:
                elegidos.pop()
            return

        # regla del pivote: próxima elección debe contener la toma mínima de rem
        pivote = min(rem)
        opciones = [i for i in disponibles if pivote in sets_cand[i] and sets_cand[i].issubset(rem)]

        # ordenar por menor costo para mejorar rápido el BEST
        opciones.sort(key=lambda i: costos[i])

        # presupuesto restante
        presupuesto = mejor_costo - costo_acum2

        # explora opciones del pivote
        for ci in opciones:
            dia = sets_cand[ci]
            # subestado tras elegir ci
            rem2 = rem - dia
            disp2 = {j for j in disponibles if sets_cand[j].isdisjoint(dia)}

            # cota inferior del subestado (r-1 días)
            lb_sub = cota_inferior_estado(rem2, disp2)

            # poda estricta por opción: c(D) + LB_sub <= presupuesto
            suma = costos[ci] + lb_sub
            if (permitir_empates_con_best and suma <= presupuesto) or (not permitir_empates_con_best and suma < presupuesto):
                elegidos.append(ci)
                dfs(rem2, disp2, elegidos, costo_acum2 + costos[ci])
                elegidos.pop()
            # si no pasa, se poda esta opción

        # deshacer los forzosos aplicados antes de salir del nodo
        for _ in extra_elegidos:
            elegidos.pop()

    # llamada inicial
    rem0 = set(todas_las_tomas)
    disp0 = set(range(N))
    dfs(rem0, disp0, [], 0)
    #dfs(rem: Set[int], disponibles: Set[int], elegidos: List[int], costo_acum: int):
    if mejor_calendario is None:
        return None, None
    return mejor_costo, [frozenset(sets_cand[i]) for i in mejor_calendario]
```
